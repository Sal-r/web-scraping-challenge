```python
#dependencies
from bs4 import BeautifulSoup as bs
import requests

#splinter
from splinter import Browser
```

```python
#code doesn't pick most recent from list, would need to update every time.
url = 'https://mars.nasa.gov/news/8657/nasas-perseverance-rover-mission-getting-in-shape-for-launch/'
```

```python
response = requests.get(url)
```

```python
soup = bs(response.text, 'html.parser')
```

```python
print(soup.prettify())
```

```python
new_article_title = soup.select('h1', class_='article_title')[0].text.strip()
```

```python
new_article_title
```

```python
paragraphs = []
new_article_para = soup.find_all('p')
for para in new_article_para:
    paragraphs.append(para.get_text(strip=True))
```

```python
paragraphs
```

```python
#article class = carousel_item - is the class for the featured image
with Browser() as browser:
    #visit url
    url = "https://www.jpl.nasa.gov/spaceimages/?search=&category=Mars"
    browser.visit(url)
    browser.fill
```
